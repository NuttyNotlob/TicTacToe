```python
#First, define everything required for the game:

import random
from IPython.display import clear_output

def print_board(marks):
    print("\n")
    print("     |     |     ")
    print(f"  {marks[0]}  |  {marks[1]}  |  {marks[2]}  ")
    print("     |     |     ")
    print("------------------") 
    print("     |     |     ")
    print(f"  {marks[3]}  |  {marks[4]}  |  {marks[5]}  ")
    print("     |     |     ")
    print("------------------")   
    print("     |     |     ")
    print(f"  {marks[6]}  |  {marks[7]}  |  {marks[8]}  ")
    print("     |     |     ")
    print("\n")
    
def board_full(marks):
    if " " in marks:
        return False
    else:
        return True
    
def complete_check_first():

    return ((marks[0] == details[first][1] and marks[1] == details[first][1] and marks[2] == details[first][1]) or 
    (marks[3] == details[first][1] and marks[4] == details[first][1] and marks[5] == details[first][1]) or
    (marks[6] == details[first][1] and marks[7] == details[first][1] and marks[8] == details[first][1]) or
    (marks[0] == details[first][1] and marks[3] == details[first][1] and marks[6] == details[first][1]) or
    (marks[1] == details[first][1] and marks[4] == details[first][1] and marks[7] == details[first][1]) or
    (marks[2] == details[first][1] and marks[5] == details[first][1] and marks[8] == details[first][1]) or
    (marks[0] == details[first][1] and marks[4] == details[first][1] and marks[8] == details[first][1]) or
    (marks[2] == details[first][1] and marks[4] == details[first][1] and marks[6] == details[first][1]))
    
def complete_check_second():
    return ((marks[0] == details[second][1] and marks[1] == details[second][1] and marks[2] == details[second][1]) or
    (marks[3] == details[second][1] and marks[4] == details[second][1] and marks[5] == details[second][1]) or
    (marks[6] == details[second][1] and marks[7] == details[second][1] and marks[8] == details[second][1]) or
    (marks[0] == details[second][1] and marks[3] == details[second][1] and marks[6] == details[second][1]) or
    (marks[1] == details[second][1] and marks[4] == details[second][1] and marks[7] == details[second][1]) or
    (marks[2] == details[second][1] and marks[5] == details[second][1] and marks[8] == details[second][1]) or
    (marks[0] == details[second][1] and marks[4] == details[second][1] and marks[8] == details[second][1]) or
    (marks[2] == details[second][1] and marks[4] == details[second][1] and marks[6] == details[second][1]))


game_play = "YES"

#Now for the game, contained in a big ol' while loop

#Set the players, and the symbols they want to use, with a check to make sure they only chose to have 
#one character as their symbol

names_okay = ""
while names_okay != "Okay":
    name1 = input("Player one, please enter your name here: ")
    symbol1 = input("\nAnd please enter the one-character symbol you'd like to use here: ")

    name2 = input("\nPlayer two, please enter your name here: ")
    symbol2 = input("\nAnd please enter the one-character symbol you'd like to use here: ")

    if len(symbol1) == 1 and len(symbol2) == 1:
        names_okay = "Okay"
    else:
        print("\n!!! Please retry, making sure you use only one character for your chosen symbols !!!")

details = {"Player1":[name1,symbol1],"Player2":[name2,symbol2]}

print("Welcome to my game of tic-tac-toe! In this game, you'll be able to choose your own symbols, \
    and play against\neach other to see who's the true winner - enjoy!\n")


while game_play == "YES":
    
#Reset the board

    marks = [" "," "," "," "," "," "," "," "," ",]

#Set who is playing first, and explain the game

    first = random.choice(list(details.keys()))

    if first == "Player1":
        second = "Player2"
    else:
        second = "Player1" 

    print(f"\n{details[first][0]}, you will go first, with the symbol of {details[first][1]}\n")

    print("Please enter a number 1 - 9, corresponding with the grid print below,\nto place your first symbol\n")

    number_grid = ["1","2","3","4","5","6","7","8","9",]

    print_board(number_grid)
    
#Now for the first turn of the game, which due to the different text,
#plays on its own in a different part of the script

    input_list = []

    user_input = input(f"{details[first][0]}, enter your first go: ")
    
    if int(user_input) in range(1,10):
        input_list.append(user_input)
        marks[int(input_list[0])-1] = f"{details[first][1]}"
        clear_output()
        print_board(marks)
    else:
        print("\nPlease make sure you pick a number between 1 and 9")
        continue
    
#Now for the remaining turns until the game is won. This includes a check to see if the game is complete,
#either by a player winning or the game going to a tie
    
    turn_number = 2
    
    while complete_check_first() != True and complete_check_second() != True:
        if turn_number % 2 == 0:
            user_input = input(f"{details[second][0]}, enter your go: ")
            if user_input in input_list:
                print("\n You can't pick a number that has previously been chosen!")
                continue
            else:
                if int(user_input) in range(1,10):
                    input_list.append(user_input)
                    marks[int(input_list[turn_number-1])-1] = f"{details[second][1]}"
                    clear_output()
                    print_board(marks)
                else:
                    print("\nPlease make sure you pick a number between 1 and 9!")
                    continue
        else:
            user_input = input(f"{details[first][0]}, enter your go: ")
            if user_input in input_list:
                print("\n You can't pick a number that has previously been chosen!")
                continue
            else:
                if int(user_input) in range(1,10):
                    input_list.append(user_input)
                    marks[int(input_list[turn_number-1])-1] = f"{details[first][1]}"
                    clear_output()
                    print_board(marks)
                else:
                    print("\nPlease make sure you pick a number between 1 and 9!")
                    continue
        turn_number+=1
        
        if board_full(marks) == True and complete_check_first() != True and complete_check_second() != True:
            print("The game is a tie!")
            break

    if complete_check_first():
        print(f"\nCongratulations {details[first][0]}, you won!\n")
    elif complete_check_second():
        print(f"\nCongratulations {details[second][0]}, you won!\n")
        
#And finally, to let players decide if they'd like to play again

    game_play = []
    while game_play != "YES":

        game_play = input("\nWould you like to play again? Please enter Yes or No: ").upper()

        if game_play == "NO":
            print("\nThanks for playing!")
            break
        if game_play != "YES" and game_play != "NO":
            print("\nPlease enter Yes or No")
            continue
```
